**2. An√°lisis estad√≠stico b√°sico**

2.1. Calcula:

     Edad promedio

     Horas promedio de sue√±o

     Nivel promedio de estr√©s

     Puntaje promedio de productividad

2.2. ¬øCu√°l es el valor m√°ximo, m√≠nimo y mediana del depression_score?

2.3. ¬øCu√°l es el rango de sleep_hours?

2.4. ¬øQu√© porcentaje del total busca tratamiento (seeks_treatment = Yes)?





```python
# 2.1. Calcula:

     # Edad promedio

     # Horas promedio de sue√±o

     # Nivel promedio de estr√©s

     # Puntaje promedio de productividad

print(f"La edad promedio es de: {round(df['age'].mean(),2)} a√±os.")

print(f"Las horas promedio de sue√±o son de: {round(df['sleep_hours'].mean(),2)} horas.")

print(f"El nivel promedio de estres es de: {round(df['stress_level'].mean(),2)}.")

print(f"El puntaje promedio de productividad es de: {round(df['productivity_score'].mean(),2)}")
```

```python
# 2.2. ¬øCu√°l es el valor m√°ximo, m√≠nimo y mediana del depression_score?

print(f"El valor m√°ximo en la puntuaci√≥n de depresi√≥n es de: {round(df['depression_score'].max(),2)}.")

print(f"El valor m√≠nimo en la puntuaci√≥n de depresi√≥n es de: {round(df['depression_score'].min(),2)}.")

print(f"El valor de la mediana en la puntuaci√≥n de depresi√≥n es de: {round(df['depression_score'].median(),2)}.")
```

```python
# 2.3. ¬øCu√°l es el rango de sleep_hours?

max = round(df['sleep_hours'].max(),2)

min = round(df['sleep_hours'].min(),2)

rango = max - min

print(f"El rango de las horas de sue√±o es de: {rango} horas.")
```

```python
# 2.4. ¬øQu√© porcentaje del total busca tratamiento (seeks_treatment = Yes)?

seeks_treatment_yes = (df['seeks_treatment'] == 'Yes').sum()
seeks_treatment_total = df['seeks_treatment'].value_counts().sum()

porcentaje = (seeks_treatment_yes / seeks_treatment_total) * 100

print(f"El porcentaje que busca trtamiento es del: {round(porcentaje,2)}%.")
```

**5. Estad√≠stica y probabilidad b√°sica**

5.1. Calcula la desviaci√≥n est√°ndar del anxiety_score.

5.2. Usa f√≥rmulas de probabilidad:
     ¬øCu√°l es la probabilidad de que una persona seleccionada al azar tenga:

     mental_health_history = Yes

     y tambi√©n seeks_treatment = Yes?

5.3. Usa √°lgebra: Calcula el √≠ndice combinado de salud mental como:

```python
# 5.1. Calcula la desviaci√≥n est√°ndar del anxiety_score.

print(f"La desviaci√≥n estandar del puntaje de ansiedad es de {round(df['anxiety_score'].std(),2)}")
```

```python
# 5.2. Usa f√≥rmulas de probabilidad:
     # ¬øCu√°l es la probabilidad de que una persona seleccionada al azar tenga:

     # mental_health_history = Yes

     # y tambi√©n seeks_treatment = Yes?

total = len(df)

conteo_condicion = len(df[(df['mental_health_history'] == 'Yes') & (df['seeks_treatment'] == 'Yes')])

probabilidad = conteo_condicion/total

print(f"La probabilidad de que esto suceda es de {probabilidad:.4f} or {probabilidad*100:.2f}%")
```

```python
# Usa √°lgebra: Calcula el √≠ndice combinado de salud mental como:
     # √≠ndice_salud_mental = (depression_score + anxiety_score) / 2

indice = (df['depression_score'] + df['anxiety_score']) / 2

print(indice)
```

1. Contar cu√°ntas personas mayores de 50 a√±os tienen mental_health_risk = High.

2. Crea una funci√≥n que reciba stress_level y retorne una categor√≠a:

     Bajo (0‚Äì3)

     Medio (4‚Äì7)

     Alto (8‚Äì10)
3. Apl√≠calo a todo el dataset y agr√©galo como una nueva columna.

```python
# 1. Contar cu√°ntas personas mayores de 50 a√±os tienen mental_health_risk = High.

conteo = len(df[(df['age'] > 50) & (df['mental_health_risk'] == 'High')])
print(f'La cantidad de personas mayores de 50 a√±os y que tienen un riesgo de salud mental alto son: {conteo}.')
print('')

#2. Crea una funci√≥n que reciba stress_level y retorne una categor√≠a:

     #Bajo (0‚Äì3)

     #Medio (4‚Äì7)

     #Alto (8‚Äì10)

def retorno_cat(columna):
    if 0 <= columna <= 3:
        return 'Bajo'
    elif 4 <= columna <= 7:
        return 'Medio'
    elif 8 <= columna <= 10:
        return 'Alto'
    
#3. Apl√≠calo a todo el dataset y agr√©galo como una nueva columna.

df['stress_category'] = df['stress_level'].apply(retorno_cat)

print(df[['stress_level','stress_category']].head())

```
