**2. Análisis estadístico básico**

2.1. Calcula:

     Edad promedio

     Horas promedio de sueño

     Nivel promedio de estrés

     Puntaje promedio de productividad

2.2. ¿Cuál es el valor máximo, mínimo y mediana del depression_score?

2.3. ¿Cuál es el rango de sleep_hours?

2.4. ¿Qué porcentaje del total busca tratamiento (seeks_treatment = Yes)?





```python
# 2.1. Calcula:

     # Edad promedio

     # Horas promedio de sueño

     # Nivel promedio de estrés

     # Puntaje promedio de productividad

print(f"La edad promedio es de: {round(df['age'].mean(),2)} años.")

print(f"Las horas promedio de sueño son de: {round(df['sleep_hours'].mean(),2)} horas.")

print(f"El nivel promedio de estres es de: {round(df['stress_level'].mean(),2)}.")

print(f"El puntaje promedio de productividad es de: {round(df['productivity_score'].mean(),2)}")
```

```python
# 2.2. ¿Cuál es el valor máximo, mínimo y mediana del depression_score?

print(f"El valor máximo en la puntuación de depresión es de: {round(df['depression_score'].max(),2)}.")

print(f"El valor mínimo en la puntuación de depresión es de: {round(df['depression_score'].min(),2)}.")

print(f"El valor de la mediana en la puntuación de depresión es de: {round(df['depression_score'].median(),2)}.")
```

```python
# 2.3. ¿Cuál es el rango de sleep_hours?

max = round(df['sleep_hours'].max(),2)

min = round(df['sleep_hours'].min(),2)

rango = max - min

print(f"El rango de las horas de sueño es de: {rango} horas.")
```

```python
# 2.4. ¿Qué porcentaje del total busca tratamiento (seeks_treatment = Yes)?

seeks_treatment_yes = (df['seeks_treatment'] == 'Yes').sum()
seeks_treatment_total = df['seeks_treatment'].value_counts().sum()

porcentaje = (seeks_treatment_yes / seeks_treatment_total) * 100

print(f"El porcentaje que busca trtamiento es del: {round(porcentaje,2)}%.")
```

**5. Estadística y probabilidad básica**

5.1. Calcula la desviación estándar del anxiety_score.

5.2. Usa fórmulas de probabilidad:
     ¿Cuál es la probabilidad de que una persona seleccionada al azar tenga:

     mental_health_history = Yes

     y también seeks_treatment = Yes?

5.3. Usa álgebra: Calcula el índice combinado de salud mental como:

```python
# 5.1. Calcula la desviación estándar del anxiety_score.

print(f"La desviación estandar del puntaje de ansiedad es de {round(df['anxiety_score'].std(),2)}")
```

```python
# 5.2. Usa fórmulas de probabilidad:
     # ¿Cuál es la probabilidad de que una persona seleccionada al azar tenga:

     # mental_health_history = Yes

     # y también seeks_treatment = Yes?

total = len(df)

conteo_condicion = len(df[(df['mental_health_history'] == 'Yes') & (df['seeks_treatment'] == 'Yes')])

probabilidad = conteo_condicion/total

print(f"La probabilidad de que esto suceda es de {probabilidad:.4f} or {probabilidad*100:.2f}%")
```

```python
# Usa álgebra: Calcula el índice combinado de salud mental como:
     # índice_salud_mental = (depression_score + anxiety_score) / 2

indice = (df['depression_score'] + df['anxiety_score']) / 2

print(indice)
```

**Extras**

1. Contar cuántas personas mayores de 50 años tienen mental_health_risk = High.

2. Crea una función que reciba stress_level y retorne una categoría:

     Bajo (0–3)

     Medio (4–7)

     Alto (8–10)
3. Aplícalo a todo el dataset y agrégalo como una nueva columna.

```python
# 1. Contar cuántas personas mayores de 50 años tienen mental_health_risk = High.

conteo = len(df[(df['age'] > 50) & (df['mental_health_risk'] == 'High')])
print(f'La cantidad de personas mayores de 50 años y que tienen un riesgo de salud mental alto son: {conteo}.')
print('')

#2. Crea una función que reciba stress_level y retorne una categoría:

     #Bajo (0–3)

     #Medio (4–7)

     #Alto (8–10)

def retorno_cat(columna):
    if 0 <= columna <= 3:
        return 'Bajo'
    elif 4 <= columna <= 7:
        return 'Medio'
    elif 8 <= columna <= 10:
        return 'Alto'
    
#3. Aplícalo a todo el dataset y agrégalo como una nueva columna.

df['stress_category'] = df['stress_level'].apply(retorno_cat)

print(df[['stress_level','stress_category']].head())

```
